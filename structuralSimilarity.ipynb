```python
from skimage.metrics import structural_similarity
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
img1path = r"D:\MS Computer Engineering\Semester 1\CV\Final Project\_\hpl-tamil-iwfhr06-train-offline\usr_22\018t07.tiff"
img2path = r"D:\MS Computer Engineering\Semester 1\CV\CVdataset\1e0fde23-6935-11ed-8af6-5c879ca2e7ab.jpg"
def strucSim(img1path, img2path):
    img1 = cv2.imread(img1path)
    img2 = cv2.imread(img2path)
    img2 = img2[int(0.25*img2.shape[0]):int(0.55*img2.shape[0]), int(0.55*img2.shape[1]):int(0.8*img2.shape[1])]
    img2 = cv2.resize(img2, dsize=[img1.shape[1], img1.shape[0]])

    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

    score, diff = structural_similarity(img1, img2, full=True)
    return score, diff, np.concatenate([img1, img2], axis=1)
score, diff, img = strucSim(img1path, img2path)
plt.imshow(img)
plt.title("Structural similarity Score: {:.3f}%".format(score * 100))
plt.show()
```
